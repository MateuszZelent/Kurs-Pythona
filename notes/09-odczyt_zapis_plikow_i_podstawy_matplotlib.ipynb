## Rozdział 1: Wprowadzenie do formatu CSV
CSV (Comma-Separated Values) to format pliku, który przechowuje dane tabelaryczne w prostym formacie tekstowym. Każdy wiersz pliku jest rekordem danych, a każdy rekord składa się z pól, które są oddzielone przecinkami.


````python

import numpy as np
import pandas as pd

# Generowanie danych
data = np.random.rand(10, 3)  # 10 wierszy, 3 kolumny
df = pd.DataFrame(data, columns=['Kolumna1', 'Kolumna2', 'Kolumna3'])

# Wyświetlenie danych
print(df)
````

## Dodatek: Ścieżki plików w systemach Windows i Linux
Ścieżka do pliku może być określona jako względna lub bezwzględna. Ścieżka bezwzględna zawiera pełne położenie pliku od korzenia systemu plików, natomiast ścieżka względna określa lokalizację pliku względem bieżącego katalogu roboczego.

### Ścieżki bezwzględne i względne w Windows
W systemie Windows ścieżka bezwzględna zaczyna się od litery dysku (np. `C:\`). Na przykład, `C:\Users\NazwaUzytkownika\plik.txt` jest ścieżką bezwzględną do pliku `plik.txt`.

Ścieżka względna w Windows nie zawiera litery dysku i zwykle zaczyna się od nazwy folderu lub `..` (kropki kropki), które oznaczają katalog nadrzędny. Na przykład, `..\Dokumenty\plik.txt` odnosi się do pliku `plik.txt` w katalogu `Dokumenty` jeden poziom wyżej względem bieżącego katalogu.


### Ścieżki bezwzględne i względne w Linux
W systemie Linux ścieżka bezwzględna zaczyna się od znaku `/` (slash), który jest korzeniem systemu plików. Przykładem może być `/home/nazwa_uzytkownika/plik.txt`.

Ścieżka względna w Linux nie zawiera początkowego slash’a i zwykle zaczyna się od nazwy folderu lub `..` (kropki kropki), które wskazują katalog nadrzędny. Na przykład, `../Dokumenty/plik.txt` odnosi się do pliku `plik.txt` w katalogu `Dokumenty`, który znajduje się jeden poziom wyżej od bieżącego katalogu.


# Przykład użycia ścieżek w Pythonie
import os

# Ścieżka bezwzględna
- abs_path = '/home/nazwa_uzytkownika/plik.txt'  # Linux
- abs_path = 'C:\Users\NazwaUzytkownika\plik.txt'  # Windows

# Ścieżka względna
rel_path = '../Dokumenty/plik.txt'  # Zakładając, że bieżący katalog jest /home/nazwa_uzytkownika/Inne

print("Pełna ścieżka bezwzględna:", os.path.abspath(rel_path))

# Funkcje pomocnicze
Istnieje kilka funkcji w module os w Pythonie, które mogą pomóc w pracy ze ścieżkami:

- os.getcwd(): Zwraca bieżący katalog roboczy.
- os.chdir(): Zmienia bieżący katalog roboczy.
- os.path.join(): Łączy ze sobą ścieżki.
- os.path.abspath(): Zwraca bezwzględną ścieżkę do pliku lub katalogu.
- os.path.normpath(): Normalizuje ścieżkę, usuwając zbędne elementy, takie jak . i ...

# Kurs NumPy: Praca z danymi w formacie CSV


## Rozdział 1: Wprowadzenie do NumPy
NumPy to jedna z podstawowych bibliotek w ekosystemie Pythona służąca do obliczeń naukowych. Zapewnia wsparcie dla dużych, wielowymiarowych tablic i macierzy, wraz z szeroką gamą funkcji matematycznych do operacji na tych tablicach.


````python
import numpy as np

# Generowanie danych
data = np.random.rand(10, 3)  # 10 wierszy, 3 kolumny

# Wyświetlenie danych
print(data)
````


## Rozdział 2: Zapis danych do pliku CSV
NumPy umożliwia zapis danych do pliku CSV za pomocą funkcji `savetxt`.


````python
np.savetxt('przykladowe_dane_numpy.csv', data, delimiter=',', header='Kolumna1,Kolumna2,Kolumna3', comments='')
````


## Rozdział 3: Wczytywanie danych z pliku CSV
Aby wczytać zapisane dane, użyjemy funkcji `loadtxt` z biblioteki NumPy.


````python
loaded_data = np.loadtxt('przykladowe_dane_numpy.csv', delimiter=',', skiprows=1)
print(loaded_data)
````

````python
# Odczytaj dane z pliku CSV, pomijając ostatni wiersz (stopkę)
naglowki = np.genfromtxt('przykladowe_dane_numpy.csv', delimiter=",", dtype=str, skip_footer=1,)

# Pobierz pierwszy wiersz
pierwszy_wiersz = naglowki[0, :] #Pierwszy wiersz, wszystkie kolumny

# Wyświetl pierwszy wiersz
print(pierwszy_wiersz)
````


## Rozdział 4: Manipulacja i przetwarzanie danych
NumPy oferuje różne funkcje do manipulacji i przetwarzania załadowanych danych, takie jak filtrowanie, selekcja kolumn, operacje matematyczne.


````python
filtered_data = loaded_data[loaded_data[:, 0] > 0.5] 
print(filtered_data)
````

## Rozdział: Praca z biblioteką `csv`
Biblioteka `csv` jest wbudowana w standardową dystrybucję Pythona i służy do czytania oraz zapisywania plików w formacie CSV (Comma Separated Values). Jest to prosty i wygodny sposób na manipulowanie danymi tabelarycznymi.


### Odczytywanie nagłówków z pliku CSV
Pierwszym krokiem w pracy z plikami CSV jest często odczytanie nagłówków, które opisują dane w każdej kolumnie.


````python
import csv

# Otwarcie pliku CSV i wczytanie nagłówków
headers = []
with open('przykladowe_dane_numpy.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader)  # Czytanie pierwszego wiersza, który zawiera nagłówki

print("Nagłówki:", headers)
````

### Odczytywanie danych z pliku CSV
Po wczytaniu nagłówków, następnym krokiem jest odczytanie rzeczywistych danych zawartych w pliku.


````python
# Wczytywanie danych z pliku CSV
data = []
with open('przykladowe_dane_numpy.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pominięcie nagłówków
    for row in reader:
        data.append(row)

print("Dane wczytane z pliku CSV:")
for row in data[:5]:  # Wyświetlenie pierwszych 5 wierszy
    print(row)
````

# Tworzenie wykresu liniowego

Aby utworzyć wykres liniowy przedstawiający średnią temperaturę w poszczególnych latach dla Polski, Niemiec i Francji, możemy użyć następującego kodu:

````python
import numpy as np
import matplotlib.pyplot as plt

# Załadowanie danych
data = np.loadtxt('przykladowe_dane_numpy.csv', delimiter=',', skiprows=1)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))  # Ustawienie rozmiaru wykresu
plt.plot(data[:, 0], label='Kolumna 1')  # Wykres dla pierwszej kolumny danych
plt.plot(data[:, 1], label='Kolumna 2')  # Wykres dla drugiej kolumny danych
plt.plot(data[:, 2], label='Kolumna 3')  # Wykres dla trzeciej kolumny danych
plt.title('Wykres danych')  # Tytuł wykresu
plt.xlabel('Indeks')  # Etykieta osi X
plt.ylabel('Wartości')  # Etykieta osi Y
plt.legend()  # Dodanie legendy
plt.show()  # Wyświetlenie wykresu
````

### Formatowanie wykresu
Dodajmy dodatkowe elementy formatujące do naszego wykresu, takie jak kolory linii, style linii oraz markery.


````python
# Formatowanie wykresu
plt.figure(figsize=(10, 6))
plt.plot(data[:, 0], 'r-', label='Kolumna 1', linewidth=2, marker='o')  # czerwona linia z okrągłymi markerami
plt.plot(data[:, 1], 'g--', label='Kolumna 2', linewidth=1, marker='s')  # zielona linia przerywana z kwadratowymi markerami
plt.plot(data[:, 2], 'b:', label='Kolumna 3', linewidth=1.5, marker='^')  # niebieska linia kropkowana z trójkątnymi markerami
plt.title('Zaawansowany wykres danych')  # Tytuł wykresu
plt.xlabel('Indeks')  # Etykieta osi X
plt.ylabel('Wartości')  # Etykieta osi Y
plt.legend()  # Dodanie legendy
plt.grid(True)  # Dodanie siatki do wykresu
plt.show()  # Wyświetlenie wykresu
````
