## Zadanie 1. Instrukcje warunkowe:

Napisz program, który poprosi użytkownika o wartość pomiędzy 0.0 a 1.0. Jeśli wartość jest poza zakresem, wypisz komunikat o błędzie. Jeśli wartość jest między 0.0 a 1.0, wypisz ocenę, korzystając z poniższej tabeli:

Wartość   Ocena
- \>= 0.9     5,0
- \>= 0.8     4,5
- \>= 0.7     4,0
- \>= 0.6     3,5
- \>= 0.5     3,0
- \< 0.5     2,0

Jeśli użytkownik wprowadzi wartość spoza zakresu, wypisz odpowiedni komunikat o błędzie i zakończ działanie programu. 

````python
def ocena_na_podstawie_wartosci():
    try:
        wartosc = float(input("Podaj wartość pomiędzy 0.0 a 1.0: "))
        
        if wartosc < 0.0 or wartosc > 1.0:
            print("Błąd: wartość poza zakresem [0.0, 1.0]")
            return
        
        if wartosc >= 0.9:
            ocena = "5,0"
        elif wartosc >= 0.8:
            ocena = "4,5"
        elif wartosc >= 0.7:
            ocena = "4,0"
        elif wartosc >= 0.6:
            ocena = "3,5"
        elif wartosc >= 0.5:
            ocena = "3,0"
        else:
            ocena = "2,0"
        
        print(f"Ocena: {ocena}")
    
    except ValueError:
        print("Błąd: wprowadzona wartość nie jest liczbą")

ocena_na_podstawie_wartosci()
````

## Zadanie 2. Funkcje:

Napisz program, który poprosi użytkownika o podanie liczby godzin oraz stawkę za godzinę pracy, a następnie używając danych wejściowych program obliczy wynagrodzenie. 

Zapłać normalną stawkę godzinową za maksymalnie 40 godzin i 1,5-krotność stawki godzinowej za wszystkie godziny przepracowane powyżej 40 godzin. 

Umieść logikę obliczenia wynagrodzenia w funkcji o nazwie computepay(), a następnie użyj tej funkcji. 

unkcja powinna zwrócić wartość. 

Powinieneś użyć input() (aby odczytać wprowadzane dane) oraz float() (aby przekonwertować wprowadzony tekst na liczbę). 

Nie przejmuj się sprawdzaniem błędów przy wprowadzaniu danych przez użytkownika - załóż, że użytkownik wpisuje liczby prawidłowo. 

Nie nazywaj swojej zmiennej o nazwie sum ani nie używaj funkcji sum().


Do przetestowania programu wykorzystaj następujące dane wejściowe:
    
liczba godzin: 45,
stawka godzinowa: 10.50.
Obliczone wynagrodzenie powinno wynosić 498.75.



````python
def computepay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * rate * 1.5
        pay = regular_pay + overtime_pay
    return pay

def main():
    hours = float(input("Podaj liczbę przepracowanych godzin: "))
    rate = float(input("Podaj stawkę godzinową: "))
    
    payment = computepay(hours, rate)
    print(f"Obliczone wynagrodzenie: {payment}")

main()
````

````python
# Część 1: Tworzenie i modyfikacja słownika
def dodaj_studentow(slownik):
    slownik["Anna"] = 4.5
    slownik["Piotr"] = 3.0
    slownik["Maria"] = 5.0
    return slownik

# Inicjalizacja słownika
oceny_studentow = {}

# Dodanie studentów
oceny_studentow = dodaj_studentow(oceny_studentow)
print("Słownik po dodaniu studentów:", oceny_studentow)

# Wypisanie oceny Anny
print(f"Ocena Anny: {oceny_studentow['Anna']}")

# Zmiana oceny Piotra
oceny_studentow["Piotr"] = 4.0
print(f"Ocena Piotra po zmianie: {oceny_studentow['Piotr']}")

# Dodanie nowego studenta
oceny_studentow["Tomek"] = 3.5
print(f"Słownik po dodaniu Tomka: {oceny_studentow}")

# Usunięcie Marii
del oceny_studentow["Maria"]
print(f"Słownik po usunięciu Marii: {oceny_studentow}")

# Część 2: Operacje na słowniku
print("\nCzęść 2:")

# Wypisanie wszystkich imion (kluczy)
print("Imiona studentów:")
for imie in oceny_studentow.keys():
    print(imie)

# Wypisanie wszystkich ocen (wartości)
print("\nOceny studentów:")
for ocena in oceny_studentow.values():
    print(ocena)

# Wypisanie par imię-ocena
print("\nPary imię-ocena:")
for imie, ocena in oceny_studentow.items():
    print(f"{imie}: {ocena}")

# Sprawdzenie czy jest student "Kasia"
kasia_ocena = oceny_studentow.get("Kasia")
if kasia_ocena is None:
    print("\nStudent 'Kasia' nie istnieje w słowniku")
else:
    print(f"\nOcena Kasi: {kasia_ocena}")

# Część 3: Liczenie wystąpień
print("\nCzęść 3:")
owoce = ["jabłko", "banan", "jabłko", "pomarańcza", "banan", "jabłko"]

licznik_owocow = {}
for owoc in owoce:
    if owoc in licznik_owocow:
        licznik_owocow[owoc] += 1
    else:
        licznik_owocow[owoc] = 1

for owoc, ilosc in licznik_owocow.items():
    print(f"{owoc}: {ilosc}")
````

# Zadanie 5: Analiza FFT sygnału sinusoidalnego

Napisz program, który generuje sygnał sinusoidalny o losowej częstotliwości z przedziału od 0 do 1000 Hz. Następnie wykonaj analizę FFT (Fast Fourier Transform) na tym sygnale i wyświetl zarówno sygnał czasowy, jak i jego widmo częstotliwościowe na wykresach za pomocą biblioteki matplotlib (jako jeden rysunek z dwoma pod-wykresami (subplot)).

````python
import numpy as np
import matplotlib.pyplot as plt

def generuj_sygnal_sinusoidalny(czas, czestotliwosc, amplituda=1.0):
    return amplituda * np.sin(2 * np.pi * czestotliwosc * czas)

def wykonaj_fft(sygnal, czestotliwosc_probkowania):
    # Wykonanie FFT
    N = len(sygnal)
    fft_wynik = np.fft.fft(sygnal)
    
    # Normalizacja widma
    fft_widmo = np.abs(fft_wynik) / N
    
    # Dla sygnału rzeczywistego widmo jest symetryczne - bierzemy tylko pierwszą połowę
    fft_widmo = fft_widmo[:N//2]
    
    # Podwojenie wszystkich wartości oprócz składowej stałej (0 Hz)
    fft_widmo[1:] *= 2
    
    # Obliczenie osi częstotliwości
    czestotliwosci = np.fft.fftfreq(N, 1/czestotliwosc_probkowania)[:N//2]
    
    return czestotliwosci, fft_widmo

# Parametry sygnału
czestotliwosc_probkowania = 10000  # Hz
czas_trwania = 1.0  # sekundy

# Generowanie losowej częstotliwości
losowa_czestotliwosc = np.random.uniform(0, 1000)  # Hz
print(f"Wygenerowana częstotliwość: {losowa_czestotliwosc:.2f} Hz")

# Generowanie osi czasu
t = np.linspace(0, czas_trwania, int(czas_trwania * czestotliwosc_probkowania), endpoint=False)

# Generowanie sygnału
sygnal = generuj_sygnal_sinusoidalny(t, losowa_czestotliwosc)

# Wykonanie FFT
czestotliwosci, widmo = wykonaj_fft(sygnal, czestotliwosc_probkowania)

# Wizualizacja
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Wykres sygnału w dziedzinie czasu
ax1.plot(t[:500], sygnal[:500])  # Pokazujemy tylko fragment sygnału dla lepszej widoczności
ax1.set_title(f'Sygnał sinusoidalny o częstotliwości {losowa_czestotliwosc:.2f} Hz')
ax1.set_xlabel('Czas [s]')
ax1.set_ylabel('Amplituda')
ax1.grid(True)

# Wykres widma częstotliwościowego
ax2.plot(czestotliwosci, widmo)
ax2.set_title('Widmo częstotliwościowe (FFT)')
ax2.set_xlabel('Częstotliwość [Hz]')
ax2.set_ylabel('Amplituda')
ax2.grid(True)

# Ograniczenie zakresu dla lepszej widoczności piku
ax2.set_xlim(0, 2 * losowa_czestotliwosc)

plt.tight_layout()
plt.show()
````
